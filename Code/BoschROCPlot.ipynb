```python
import pandas as pd
% matplotlib inline
import matplotlib.pyplot as plt
```

### Test data

```python
viya = pd.read_csv('viyatest.csv')

testdown= pd.read_csv('testorigind.csv')
xgboostfpro=testdown['fprval']
xgboosttpro=testdown['tprval']

randomforestr = pd.read_csv('testrandomforest.csv')
randomforestrfpro=randomforestr['fprval']
randomforestrtpro=randomforestr['tprval']
```

```python
testsvm = pd.read_csv('testsvm.csv')
svmfpro=testsvm['fprval']
svmtpro=testsvm['tprval']
```

```python
def tpr(x):    
    return x['_TP_'] / (x['_TP_']+x['_FN_'])  

def fpr(x):
    return x['_FP_']/(x['_FP_']+x['_TN_'])

tprviya = viya.apply(tpr, axis=1) 
fprviya = viya.apply(fpr,axis=1)
```

```python
viyaq = pd.read_csv('qviyatest.csv')

tprqviyaq = viyaq.apply(tpr, axis=1) 
fprqviyaq = viyaq.apply(fpr, axis=1)
```

```python
pltc=list(map(lambda x:x/100.0, range(1,100)))
```

```python
#plt.plot(fprval,tprval,'r-', label='validation dataset')

plt.plot(fprviya, tprviya, 'r-', label='SAS Viya GBM')
plt.plot(xgboostfpro, xgboosttpro, 'g-', label='Xgboost')
plt.plot(randomforestrfpro, randomforestrtpro, 'b-', label='Random Forest')
plt.plot(svmfpro, svmtpro, 'k-', label='SVM')
plt.plot(fprqviyaq, tprqviyaq, 'y-', label='Quantitle SAS Viya GBM')
#plt.plot(pltc,pltc,'y-')
plt.legend(loc=4)
plt.ylabel('TPR')
plt.xlabel('FPR') 
plt.title('ROC curve on test data')
```
